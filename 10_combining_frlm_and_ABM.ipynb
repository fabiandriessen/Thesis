# Process results ABM runs

```python
import pickle
import numpy as np
from os import walk
f = []
for (dirpath, dirnames, filenames) in walk('ABM/own_work/results_without/'):
    f.extend(filenames)
    break
```

```python
all_cs_data = {'r':[], 'c':[], 'm':[], 'stations_per_loc':[], 'avg_occ':[], 'avg_line':[], 'avg_max_line':[]}
all_cs_data_full = {'r':[], 'c':[], 'm':[], 'stations_per_loc':[], 'avg_occ':[], 'avg_line':[], 'avg_max_line':[]}
all_vessel_data = {'r':[], 'c':[], 'm':[], 'supported_routes':[], 'avg_travel':[], 'avg_inline':[], 'avg_charging':[],
                   'frac_charging':[], 'frac_inline':[], 'frac_sailing':[], 'battery_size':[], 'total_charged':[]}
```

```python
df_cs_all = pd.DataFrame()
```

```python
for filename in f:
    r = int(filename[16:19].replace('.',''))
    m = int(filename.split('.')[1][-1])
    # m=3
    c = int(10000/m)

    df = pickle.load(open('ABM/own_work/results_without/'+str(filename), 'rb'))
    df = pd.DataFrame(df)
    df_cs = get_cs_data_batch(df)
    df_vessels = get_vessel_data_batch(df)

    df_cs['c'] = c
    df_cs['r'] = r
    df_cs['m'] = m

    df_cs_all =  pd.concat([df_cs_all, df_cs])

    all_cs_data['avg_occ'].append(df_cs.occupation.mean())
    all_cs_data['avg_max_line'].append(df_cs.max_line.mean())
    all_cs_data['avg_line'].append(df_cs.avg_line.mean())
    all_cs_data['stations_per_loc'].append(df_cs.charging_stations.mean())

    df_vessels['frac_inline'] = df_vessels.apply(lambda x: x.time_in_line / (x.travel_time + x.time_in_line + x.time_charging), axis=1)
    df_vessels['frac_charging'] = df_vessels.apply(lambda x: x.time_charging / (x.travel_time + x.time_in_line + x.time_charging), axis=1)
    df_vessels['frac_not_sailing'] = df_vessels['frac_charging'] + df_vessels['frac_inline']
    df_vessels['frac_sailing'] = 1 - df_vessels['frac_not_sailing']

    df_vessels_mean = df_vessels.groupby('route').mean()

    all_vessel_data['supported_routes'].append(len(df_vessels_mean))
    all_vessel_data['avg_travel'].append(df_vessels_mean.travel_time.mean())
    all_vessel_data['avg_charging'].append(df_vessels_mean.time_charging.mean())
    all_vessel_data['avg_inline'].append(df_vessels_mean.time_in_line.mean())
    all_vessel_data['frac_charging'].append(df_vessels_mean.frac_charging.mean())
    all_vessel_data['frac_inline'].append(df_vessels_mean.frac_inline.mean())
    all_vessel_data['frac_sailing'].append(df_vessels_mean.frac_sailing.mean())
    all_vessel_data['battery_size'].append(df_vessels_mean.battery_size.mean())
    all_vessel_data['total_charged'].append(df_vessels.total_charged)


    all_vessel_data['r'].append(r)
    all_vessel_data['c'].append(c)
    all_vessel_data['m'].append(m)
    all_cs_data['r'].append(r)
    all_cs_data['c'].append(c)
    all_cs_data['m'].append(m)
```

```python
cs_df = pd.DataFrame(all_cs_data)
vessel_df = pd.DataFrame(all_vessel_data)
```
