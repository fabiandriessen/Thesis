```python
def create_key(o, d, r_v):
    key1 = (o, d, r_v)
    return key1
```

```python
df_ivs['key'] = df_ivs.apply(lambda x: create_key(x.origin, x.destination, x.route_v), axis=1)
```

```python
supported_routes = list(df_vessels.route.unique())
```

```python
supported_routes
```

```python
#subset supported routes here before we take sum!
df_ivs = df_ivs.loc[df_ivs.key.isin(supported_routes)]
```

```python
df_ivs_group = df_ivs.groupby("hour").sum()
```

```python
df_ivs_group
```

```python
df_ivs = pickle.load(open("data/revised_cleaning_results/ivs_exploded_100.p", "rb"))
fig, ax = plt.subplots(1, 2, dpi=200, figsize=(12,6))
sns.barplot(x=df_hourp.index, y=df_hourp.id/(100*7), ax=ax[0])
ax[0].set_ylabel('Trips per hour [#]')
ax[0].set_xlabel('Hour of departure')
ax[0].set_title('Generated data average cumulative trips in a week')

sns.barplot(x=df_ivs_group.index, y=df_ivs_group.trip_count/(52*7), ax=ax[1])
ax[1].set_ylabel('Trips per hour [#]')
ax[1].set_xlabel('Hour of departure')
ax[1].set_title('Empirical data average cumulative trips in a week')
plt.show()
```
